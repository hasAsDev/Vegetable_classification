```python
# epoch-accuracy
import matplotlib
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd

train = pd.read_csv('run-VGG16_Model-20240813-021615_train-tag-epoch_loss.csv')
val = pd.read_csv('run-VGG16_Model-20240813-021615_validation-tag-epoch_loss.csv')
train_epo = train['Step']
train_acc = train['Value']
val_epo = val['Step']
val_acc = val['Value']

fig, ax = plt.subplots()

ax.plot(train_epo, train_acc, label='Training accuracy')
ax.plot(val_epo, val_acc, label='Validation accuracy')

ax.set_xlabel('Epoch')
ax.set_ylabel('Accracy')

ax.legend()
ax.grid(c='white')
ax.set_facecolor('#EAEAF2')

plt.show()

# fig.savefig('ResNet50_Model-epoch_accuracy.png')
```

```python
# epoch-loss
import matplotlib
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd

train = pd.read_csv('run-VGG16_Model-20240813-021615_train-tag-epoch_loss.csv')
val = pd.read_csv('run-VGG16_Model-20240813-021615_validation-tag-epoch_loss.csv')
train_epo = train['Step']
train_acc = train['Value']
val_epo = val['Step']
val_acc = val['Value']

fig, ax = plt.subplots()

ax.plot(train_epo, train_acc, label='Training loss')
ax.plot(val_epo, val_acc, label='Validation loss')

ax.set_xlabel('Epoch')
ax.set_ylabel('Loss')

ax.legend()
ax.grid(c='white')
ax.set_facecolor('#EAEAF2')

plt.show()
```

```python
# visualize data
import cv2 as cv
import matplotlib.pyplot as plt
import numpy as np
import tensorflow as tf
import keras
import os
# Load dataset
# get data_dir for tranining
current_path = os.getcwd()

img_height = 224
img_width = 224
batch_size = 1000

# set train dataset
train_data_dir = os.path.join(current_path, '..\Vegetable_Images\\train')
train_ds = keras.utils.image_dataset_from_directory(
    train_data_dir,
    label_mode='categorical',
    image_size = (img_height, img_width),
    batch_size = batch_size,
    shuffle=False
)

class_names = train_ds.class_names
class_num = len(class_names)


plt.figure(figsize=(10, 10))
for i in range(15):
    ax = plt.subplot(3, 5, i + 1)
    for images, labels in train_ds.take(i+1):
        plt.imshow(images[0].numpy().astype("uint8"))
        plt.title(class_names[np.argmax(labels[i])])
        plt.axis("off")
```
