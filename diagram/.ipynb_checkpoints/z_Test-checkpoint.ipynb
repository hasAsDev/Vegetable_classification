```python
import cv2 as cv
import matplotlib.pyplot as plt
import numpy as np
import tensorflow as tf
import keras
import os
# Load dataset
# get data_dir for tranining
current_path = os.getcwd()

img_height = 224
img_width = 224
batch_size = 128

# set test dataset
test_data_dir = os.path.join(current_path, 'Vegetable_Images\\test')
test_ds = keras.utils.image_dataset_from_directory(
    test_data_dir,
    label_mode='categorical',
    image_size = (img_height, img_width),
    batch_size = batch_size
)

class_names = test_ds.class_names
class_num = len(class_names)
```

```python
# Predict
import cv2 as cv
import matplotlib.pyplot as plt
import numpy as np
import tensorflow as tf
import keras
# load model
model = tf.keras.models.load_model('saved_models\BfS_Model.keras')
# load image
img = cv.imread('Vegetable_Images\\test\\Cauliflower\\1196.jpg')[...,::-1]

def predict(model, img):
    class_names = np.array(
        ['Bean', 'Bitter_Gourd', 'Bottle_Gourd', 'Brinjal', 'Broccoli', 'Cabbage', 'Capsicum', 
         'Carrot', 'Cauliflower', 'Cucumber', 'Papaya', 'Potato', 'Pumpkin', 'Radish', 'Tomato']
    )
    # ensure img right to input shape
    img = img_array = tf.keras.utils.img_to_array(img)
    img_array = tf.expand_dims(img, 0)

    # predict
    predictions = model.predict(img_array)
    class_predicted = class_names[np.argmax(predictions)]
    return predictions, class_predicted, img_array
    

predictions, class_predicted, img_array = predict(model, img)

print(f'Input shape: {img_array.shape}')
print(f'Output shape: {predictions.shape}')
print(f'Output : {predictions}')
print(f'Predicted Class: {class_predicted}')
plt.imshow(img)
```

```python
predicted_label = np.argmax(model.predict(test_ds), axis=1)
predicted_label[class_names]
```

```python
# confusion matrix
from sklearn.metrics import confusion_matrix, ConfusionMatrixDisplay



cfd = ConfusionMatrixDisplay.from_predictions(
    y_true=['Bean', 'Bitter_Gourd', 'Bottle_Gourd', 'Bottle_Gourd'], 
    y_pred=['Bean', 'Bitter_Gourd', 'Bottle_Gourd', 'Bottle_Gourd'],
    labels=class_names,
    cmap='Blues',
    ax=plt.subplots(figsize=(15, 10))[1],
    xticks_rotation='vertical'
)

cfd.figure_.savefig('diagram/test.png', pad_inches=3.0)
```
